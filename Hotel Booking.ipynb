```python
def conditions(df):
    if (df['children'] >= 1) or (df['babies'] >= 1):
        if (df['adults'] >= 1):
            return 'family'
    else:
        if (df['adults'] == 2):
            return 'couple'
        else:
            return 'single or others'
df['guest_type'] = df.apply(conditions, axis=1)

s = df['guest_type'].value_counts().values.sum()

def fmt(x):
    return '{:.1f}%\n{:.0f}'.format(x, s*x/100)
plt.pie(df['guest_type'].value_counts().values, labels=df['guest_type'].value_counts().index, autopct=fmt)
plt.title('Types of Guest in Both Hotel')
plt.show()
```
